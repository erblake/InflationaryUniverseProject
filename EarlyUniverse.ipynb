```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.axes as ax
import scipy
from scipy import constants
from scipy.signal import find_peaks
import pandas as pd
from scipy.signal import argrelextrema

x = np.arange(0.1, 5*np.pi, 0.1)
y = np.abs(-1+np.log(x**-2))
y2 =  12*np.exp(-x)*np.sinh(np.sqrt(1-1/125)*x)


plt.plot(x, y, color='green', label='Exponential Potential')
plt.plot(x, y2, color='red', label='Power-Law Potential')
plt.ylabel(r'$|\delta\phi/\phi_0|$ (a.u.)')
plt.xlabel('Time (a.u.)') 
plt.legend()
plt.show()
```
